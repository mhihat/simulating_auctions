```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
```

```python
N=10000

x = np.linspace(0.5,3,100)
y = []

for alpha in x :
    expec = 0    
    for _ in range(N) :
        v1 = rand.rand()*alpha
        v2 = rand.rand()
        expec += min(v1,v2)-max(v1,v2)/2
    y.append(expec/N)
    
    
theory = [-(al-1)**2/(4*al) for al in x]

plt.plot(x,y)
plt.plot(x,theory,'r')
plt.axhline(0)
```

```python
N=100000

x = np.linspace(0,20,100)
Y = []

for alpha in x :
    expec = 0    
    for _ in range(N) :
        v1 = rand.rand()*alpha
        v2 = rand.rand()
        y = rand.rand()

        mini = min(v1,v2)
        maxi = max(v1,v2)
        b = min(maxi/2,1)

        if(b>=y) :
            expec += b-max(mini,2*y)
    Y.append(expec/N)
```

```python
plt.figure(figsize=(10,5))

plt.plot(np.linspace(0.1,3.1,100),Y,'r',label=r'$\sum m_i(0)$')
#plt.plot(x[15:],theory(x[15:]),label='theory')

plt.title(r"Evolution de $\sum m_i(0)$, pour $G=F_2=Unif[0,1], \ F_1=Unif[0,\omega_1]$ en fonction de $\omega_1$")
plt.legend()
plt.xlabel(r"$\omega_1$")
plt.axhline(0)
plt.savefig("sum mi")
```
