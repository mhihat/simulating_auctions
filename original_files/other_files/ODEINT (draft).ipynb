```python
import numpy as np
from scipy import integrate, stats
```

```python
n=2

F_ = [stats.uniform.cdf for _ in range(n)]
f_ = [stats.uniform.pdf for _ in range(n)]
t = np.linspace(0.01,1,101)

def inverse_edo(y,t) :
    return np.array([F_[i](y[i])/f_[i](y[i])*(np.sum(1/(y-t))/(n-1)-1/(y[i]-t)) for i in range(n)])

y0 = np.zeros(n)
```

```python
integrate.odeint(inverse_edo,y0,t)
```
